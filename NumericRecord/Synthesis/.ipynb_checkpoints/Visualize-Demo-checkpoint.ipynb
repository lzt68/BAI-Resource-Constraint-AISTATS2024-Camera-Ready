## L=2

```python
# plot the result with error bar, K=256, C=1500, 1500, only SHRR_his
%reset -f
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

filefolder = "./L_2/"

barlabel_fontsize = 32
legend_fontsize = 36
label_fontsize = 36
axis_fontsize = 36
title_fontsize = 44

data_uniform = pd.read_csv(filefolder + "Uniform_Agent.csv")
# data_SH = pd.read_csv(filefolder + "SequentialHalvingRR_Recycle_FailureFlag_Agent.csv")
data_SH_history = pd.read_csv(filefolder + "SequentialHalvingRR_Recycle_FailureFlag_History_Agent.csv")
data_ucb = pd.read_csv(filefolder + "UCB_Agent.csv")
data_AT_lucb = pd.read_csv(filefolder + "AT_LUCB_Agent.csv")
data_DSH = pd.read_csv(filefolder + "DoublingSequentialHalving_Agent.csv")

env_replace_value_dict = {
    # "Env_FixedConsumption": "Fixed", 
    "Env_Uncorrelated_Reward": "Uncorrelated",
    "Env_Correlated_Uniform": "Correlated",
    
    "Uniform_Agent": "Uniform",
    # "SequentialHalvingRR_Recycle_FailureFlag_Agent": "SHRR",
    "SequentialHalvingRR_Recycle_FailureFlag_History_Agent": "SHRR",
    "UCB_Agent": "UCB",
    "AT_LUCB_Agent": "AT-LUCB",
    "DoublingSequentialHalving_Agent": "DSH"
}

data_uniform.replace(env_replace_value_dict, inplace=True)
# data_SH.replace(env_replace_value_dict, inplace=True)
data_ucb.replace(env_replace_value_dict, inplace=True)
data_AT_lucb.replace(env_replace_value_dict, inplace=True)
data_DSH.replace(env_replace_value_dict, inplace=True)
data_SH_history.replace(env_replace_value_dict, inplace=True)

C = "[1500. 1500.]"
K = 256
para_name_list = [
    "One_suboptimal",
    "Balanced_Trap_HalfConsumption_Fix32",
    "Poly",
    "Geometry",
]
ncols = 2
nrows = (len(para_name_list)+1) // ncols
# ncols = 4
# nrows = 1

# begin to plot
fig, ax_ = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 10, 8*nrows), sharey=True)

for ii, para_name in enumerate(para_name_list):
    ax = ax_[ii//ncols][ii%ncols]
    # ax = ax_[0][ii]
    # ax = ax_[ii]
    
    para_type = para_name + "_MultiR"
    # data_concat = pd.concat([data_SH, data_SH_history, data_AT_lucb, data_ucb, data_uniform, data_DSH], axis=0)
    data_concat = pd.concat([data_SH_history, data_AT_lucb, data_ucb, data_uniform, data_DSH], axis=0)
    
    ## sift the data
    data_concat = data_concat[
        (data_concat["K"]==K) &
        ((data_concat["C"]==C)) & 
        ((data_concat["para-type"] == para_type + "_HH") | (data_concat["para-type"] == para_type + "_HL") | (data_concat["para-type"] == para_type + "_Mixture")) &
        (data_concat["env"] != "Fixed")
    ]
    data_concat["Failure_Prob"] = 1 - data_concat["success"]
    ## modify the label
    data_concat["para-type"] = data_concat["para-type"].str.replace(
        para_type, 
        ""
    ).replace(
        "_HH", 
        "High match High,"
    ).replace(
        "_HL", 
        "High match Low,"
    ).replace(
        "_Mixture",
        "Mixture,"
    )
    data_concat["exp_setting"] =(
        data_concat["para-type"] + " " + 
        data_concat["env"]
    )
    # order = ["SHRR", "SHRR_his", "AT-LUCB", "UCB", "Uniform", "DSH"]
    order = ["SHRR", "AT-LUCB", "UCB", "Uniform", "DSH"]
    data_concat["agent"] = pd.CategoricalIndex(data_concat["agent"], ordered=True, categories=order)
    order = [ 
        "High match High, Correlated", 
        "Mixture, Correlated",
        "High match Low, Correlated", 
        "High match High, Uncorrelated", 
        "Mixture, Uncorrelated",
        "High match Low, Uncorrelated",
    ]
    data_concat["exp_setting"] = pd.CategoricalIndex(data_concat["exp_setting"], ordered=True, categories=order)
    data_concat = data_concat.sort_values(["exp_setting", "agent"])
    data_concat = data_concat[["agent", "exp_setting", "Failure_Prob", "std_success_rate"]]
    
    ## plot the figure
    print(data_concat.shape) # it should be (30, 3)
    g = sns.barplot(data=data_concat, x="agent", y="Failure_Prob", hue="exp_setting", ci=None, dodge=True, ax=ax)
    # ax.set_xlabel("Agent", fontsize=label_fontsize)
    ax.set_xlabel("", fontsize=label_fontsize)
    ax.set_xticklabels(g.get_xticklabels(), rotation=90)
    if ii % ncols == 0:
        ax.set_ylabel("Failure Probability", fontsize=label_fontsize)
    else:
        ax.set_ylabel("", fontsize=label_fontsize)
    ax.set_ylim(bottom=0., top=1.)
    if para_name == "Balanced_Trap_HalfConsumption_Fix32":
        ax.set_title("Trap", fontsize=title_fontsize)
    elif para_name == "Linear":
        ax.set_title("Arithmetic", fontsize=title_fontsize)
    else:
        ax.set_title(para_name, fontsize=title_fontsize)
    ax.tick_params(axis='both', which='major', labelsize=axis_fontsize)
    ax.legend([],[], frameon=False)
    x_coords = [p.get_x() + 0.5 * p.get_width() for p in g.patches]
    y_coords = [p.get_height() for p in g.patches]
    g.errorbar(x=x_coords, y=y_coords, yerr=data_concat["std_success_rate"], fmt="none", c="k")
    
# add legend
handles, labels = ax_[0][0].get_legend_handles_labels()
# handles, labels = ax_[0].get_legend_handles_labels()
# fig.legend(handles, labels, loc='lower center', fontsize=legend_fontsize, bbox_to_anchor=(0.5, -0.05))
plt.tight_layout()
fig.legend(handles, labels, loc='upper center', fontsize=legend_fontsize, ncol=2, bbox_to_anchor=[0.5, 0.])
fig.suptitle(f"K={K}, C={C}, L=2", fontsize=title_fontsize, y=1.05)
plt.savefig("./K-256_C-1500-1500.eps", bbox_inches='tight')
plt.show()
```
